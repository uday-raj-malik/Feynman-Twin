````
import os
from pypdf import PdfReader

base = '/content/drive/MyDrive/feynman-twin/data'

documents = []  # Will store {"text": "...", "source": "filename"}

for root, dirs, files in os.walk(base):
    for file in files:
        path = os.path.join(root, file)
        source = file

        try:
            # Handle PDFs
            if file.endswith('.pdf'):
                reader = PdfReader(path)
                text = ""
                for page in reader.pages:
                    text += page.extract_text() or ""
                documents.append({"text": text, "source": source})
                print(f'✅ PDF loaded: {file} ({len(text)} chars)')

            # Handle text files
            elif file.endswith('.txt'):
                with open(path, 'r', encoding='utf-8', errors='ignore') as f:
                    text = f.read()
                documents.append({"text": text, "source": source})
                print(f'✅ TXT loaded: {file} ({len(text)} chars)')

        except Exception as e:
            print(f'❌ Failed: {file} — {e}')

print(f'\n📚 Total documents loaded: {len(documents)}')
total_chars = sum(len(d["text"]) for d in documents)
print(f'📝 Total characters: {total_chars:,}')
````

````
def chunk_text(text, source, chunk_size=800, overlap=100):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        if len(chunk.strip()) > 100:  # skip tiny chunks
            chunks.append({"text": chunk.strip(), "source": source})
        start += chunk_size - overlap  # overlap keeps context between chunks
    return chunks

# Chunk all documents
all_chunks = []

for doc in documents:
    if len(doc["text"]) < 200:  # skip nearly empty docs
        print(f'⏭️  Skipping (too short): {doc["source"]}')
        continue
    chunks = chunk_text(doc["text"], doc["source"])
    all_chunks.append(chunks)
    print(f'✅ {doc["source"]} → {len(chunks)} chunks')

# Flatten
all_chunks = [chunk for sublist in all_chunks for chunk in sublist]

print(f'\n📦 Total chunks ready: {len(all_chunks)}')
print(f'📄 Sample chunk:\n{"-"*40}\n{all_chunks[50]["text"][:300]}\n{"-"*40}')
````

````
import chromadb
import google.generativeai as genai
from sentence_transformers import SentenceTransformer
import time

# Load the local embedding model (downloads once, ~90MB)
print("📥 Loading embedding model...")
embedder = SentenceTransformer('all-MiniLM-L6-v2')
print("✅ Model loaded!\n")

# Set up ChromaDB
chroma_client = chromadb.PersistentClient(
    path="/content/drive/MyDrive/feynman-twin/chroma_db"
)

try:
    chroma_client.delete_collection("feynman")
    print("🗑️ Cleared old collection")
except:
    pass

collection = chroma_client.create_collection(
    name="feynman",
    metadata={"hnsw:space": "cosine"}
)

print(f"📦 Embedding {len(all_chunks)} chunks locally — no rate limits!\n")

BATCH_SIZE = 100  # can go fast now, no API limits
success_count = 0

for i in range(0, len(all_chunks), BATCH_SIZE):
    batch = all_chunks[i:i+BATCH_SIZE]

    texts = [c["text"] for c in batch]
    sources = [c["source"] for c in batch]
    ids = [str(i + j) for j in range(len(batch))]

    # Embed locally
    embeddings = embedder.encode(texts, show_progress_bar=False).tolist()

    collection.add(
        ids=ids,
        embeddings=embeddings,
        documents=texts,
        metadatas=[{"source": s} for s in sources]
    )

    success_count += len(batch)
    print(f"⏳ Progress: {success_count}/{len(all_chunks)} chunks embedded...")

print(f"\n✅ Done! {success_count} chunks embedded")
print(f"📊 Collection size: {collection.count()} vectors")
````

````
def retrieve(query, top_k=5):
    # Embed the question using the same local model
    query_embedding = embedder.encode([query]).tolist()[0]

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )

    chunks = results["documents"][0]
    sources = [m["source"] for m in results["metadatas"][0]]

    return chunks, sources

# Test with 3 different queries
test_queries = [
    "What are Feynman diagrams?",
    "How do you explain quantum mechanics simply?",
    "What is the pleasure of finding things out?"
]

for query in test_queries:
    print(f"\n🔍 Query: {query}")
    print("-" * 50)
    chunks, sources = retrieve(query)
    print(f"📄 Source: {sources[0]}")
    print(f"📝 Retrieved: {chunks[0][:300]}...")
````

````
import google.generativeai as genai_old
import json
import os
import time
import re

FEYNMAN_PROMPT = """You are Richard P. Feynman — Nobel Prize-winning physicist (1965),
professor at Caltech, bongo drummer, safe-cracker, and one of the greatest
science communicators who ever lived. You died in February 1988.

== YOUR PERSONALITY ==
- You are infectiously curious. You find genuine joy in figuring things out.
- You deeply distrust pompous language and unnecessary jargon.
- You are playful and mischievous — you love jokes and personal stories.
- You are brutally honest. You say "I don't know" when you don't know.
- You believe the pleasure of finding things out is life's greatest joy.

== HOW YOU SPEAK ==
- You say things like "You see...", "Now here's the thing...",
  "What do we actually mean by that?", "Let me put it this way..."
- You build understanding from FIRST PRINCIPLES, not memorized facts.
- You use analogies from everyday life constantly.
- You get visibly excited when a question is genuinely interesting.
- You occasionally go on tangents before returning to the point.

== YOUR EXPERTISE ==
- Quantum electrodynamics (QED) — your Nobel Prize work
- Path integral formulation of quantum mechanics
- The Feynman Lectures on Physics
- Nanotechnology ("There's Plenty of Room at the Bottom")
- Superfluidity and liquid helium
- The Challenger disaster investigation (1986)
- Safe-cracking, bongo drums, art

== STRICT RULES ==
- NEVER break character. You are always Feynman.
- If asked about events after 1988, say you don't know.
- When using retrieved context, speak it naturally as your own thoughts.
- Never say "As an AI" or anything that breaks the illusion.
- If you don't know something, say so with curiosity.

== WHAT YOU KNOW ABOUT THIS PERSON ==
{long_term_memory}
"""

MEMORY_FILE = '/content/drive/MyDrive/feynman-twin/memory/long_term_memory.json'

def load_memory():
    if os.path.exists(MEMORY_FILE):
        with open(MEMORY_FILE, 'r') as f:
            return json.load(f)
    return {
        "user_name": None,
        "user_background": None,
        "topics_discussed": [],
        "interesting_facts_about_user": []
    }

def save_memory(memory):
    os.makedirs(os.path.dirname(MEMORY_FILE), exist_ok=True)
    with open(MEMORY_FILE, 'w') as f:
        json.dump(memory, f, indent=2)

def format_memory(memory):
    lines = []
    if memory.get("user_name"):
        lines.append(f"- Their name is {memory['user_name']}.")
    if memory.get("user_background"):
        lines.append(f"- Background: {memory['user_background']}.")
    if memory.get("topics_discussed"):
        topics = ", ".join(memory["topics_discussed"][-5:])
        lines.append(f"- We've talked about: {topics}.")
    if memory.get("interesting_facts_about_user"):
        for fact in memory["interesting_facts_about_user"][-2:]:
            lines.append(f"- {fact}")
    return "\n".join(lines) if lines else "I don't know this person yet."

def extract_name_directly(user_message):
    """Regex-based name extractor — zero API calls, always reliable."""
    patterns = [
        r"my name is (\w+)",
        r"i(?:'m| am) (\w+)",
        r"call me (\w+)",
        r"name'?s? (\w+)",
        r"actually (?:my name is )?(\w+)",
        r"it(?:'s| is) (\w+)",
    ]
    for pattern in patterns:
        match = re.search(pattern, user_message.lower())
        if match:
            name = match.group(1).capitalize()
            # filter out common false positives
            if name.lower() not in ["a", "an", "the", "not", "just", "still", "actually"]:
                return name
    return None


class FeynmanTwin:

    def __init__(self):
        self.long_memory = load_memory()
        self.short_memory = []
        self.model = genai_old.GenerativeModel(
            model_name="gemini-2.5-flash",
            system_instruction=FEYNMAN_PROMPT.format(
                long_term_memory=format_memory(self.long_memory)
            ),
            generation_config=genai_old.GenerationConfig(temperature=0.8)
        )
        print("✅ Feynman Twin ready!")

    def refresh_system_prompt(self):
        """Call this after memory updates so Feynman knows the latest info."""
        self.model = genai_old.GenerativeModel(
            model_name="gemini-2.5-flash",
            system_instruction=FEYNMAN_PROMPT.format(
                long_term_memory=format_memory(self.long_memory)
            ),
            generation_config=genai_old.GenerationConfig(temperature=0.8)
        )

    def chat(self, user_message):
        # 1. Retrieve relevant context from RAG
        chunks, sources = retrieve(user_message, top_k=4)
        context = "\n\n---\n\n".join(chunks)

        # 2. Augment the user message with retrieved context
        augmented_message = f"""Relevant excerpts from your writings:
{context}
---
User: {user_message}"""

        # 3. Add to history
        self.short_memory.append({
            "role": "user",
            "parts": [augmented_message]
        })

        # 4. Call Gemini
        response = self.model.generate_content(self.short_memory)
        reply = response.text

        # 5. Save reply to history
        self.short_memory.append({
            "role": "model",
            "parts": [reply]
        })

        # 6. Keep only last 20 turns to avoid token overflow
        if len(self.short_memory) > 20:
            self.short_memory = self.short_memory[-20:]

        return reply

    def update_memory(self, key, value):
        self.long_memory[key] = value
        save_memory(self.long_memory)
        self.refresh_system_prompt()
        print(f"🧠 Memory updated: {key} = {value}")


def extract_and_save_memory(feynman_instance, user_message):
    # 1. Regex name extraction first — zero API calls
    name = extract_name_directly(user_message)
    if name:
        feynman_instance.long_memory["user_name"] = name
        save_memory(feynman_instance.long_memory)
        feynman_instance.refresh_system_prompt()  # so Feynman uses correct name immediately
        print(f"🧠 Name updated to: {name}")
        return  # skip LLM call entirely

    # 2. LLM extraction only for non-name facts
    extraction_prompt = f"""
From this user message, extract any personal facts about the USER.
Return JSON only, no markdown, no extra text:
{{
  "user_name": null or string,
  "user_background": null or string,
  "new_fact": null or string
}}
User message: {user_message}
"""
    try:
        r = genai_old.GenerativeModel("gemini-2.5-flash").generate_content(extraction_prompt)
        text = r.text.replace("```json", "").replace("```", "").strip()
        facts = json.loads(text)
        if facts.get("user_name"):
            feynman_instance.long_memory["user_name"] = facts["user_name"]
        if facts.get("user_background"):
            feynman_instance.long_memory["user_background"] = facts["user_background"]
        if facts.get("new_fact"):
            feynman_instance.long_memory["interesting_facts_about_user"].append(facts["new_fact"])
        save_memory(feynman_instance.long_memory)
        feynman_instance.refresh_system_prompt()
        print(f"🧠 Memory saved: {facts}")
    except Exception as e:
        print(f"⚠️ Memory extraction failed: {e}")


def safe_chat(feynman_instance, message, retries=3):
    for attempt in range(retries):
        try:
            return feynman_instance.chat(message)
        except Exception as e:
            if "429" in str(e) or "503" in str(e) or "UNAVAILABLE" in str(e):
                wait = 30 * (attempt + 1)
                print(f"⏳ Rate limited, waiting {wait}s... (attempt {attempt+1}/{retries})")
                time.sleep(wait)
            else:
                raise e
    print("❌ Failed after all retries")
    return None


# ── FIX MEMORY FILE DIRECTLY ──────────────────────────────
with open(MEMORY_FILE, 'r') as f:
    mem = json.load(f)
mem["user_name"] = "Uday"
with open(MEMORY_FILE, 'w') as f:
    json.dump(mem, f, indent=2)
print(f"✅ Memory fixed: {mem}")

# ── TEST ──────────────────────────────────────────────────
feynman = FeynmanTwin()

msg1 = "My name is Uday and I'm a physics student from Delhi."
reply1 = safe_chat(feynman, msg1)
extract_and_save_memory(feynman, msg1)
print(reply1)

time.sleep(30)

msg2 = "What did I just tell you my name was?"
reply2 = safe_chat(feynman, msg2)
print(reply2)
````

````
import time
import json

def safe_chat(feynman_instance, message, retries=3):
    for attempt in range(retries):
        try:
            return feynman_instance.chat(message)
        except Exception as e:
            if "503" in str(e) or "UNAVAILABLE" in str(e) or "429" in str(e):
                print(f"⏳ Server busy or quota exceeded, waiting 10 seconds... (attempt {attempt+1}/{retries})")
                time.sleep(10)
            else:
                raise e
    print("❌ Failed after all retries")
    return None

def extract_and_save_memory(feynman_instance, user_message):
    extraction_prompt = f"""
From this user message, extract any personal facts about the USER.
Return JSON only, no markdown, no extra text:
{{
  "user_name": null or string,
  "user_background": null or string,
  "new_fact": null or string
}}

User message: {user_message}
"""
    try:
        r = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=extraction_prompt
        )
        text = r.text.replace("```json", "").replace("```", "").strip()
        facts = json.loads(text)
        if facts.get("user_name"):
            feynman_instance.long_memory["user_name"] = facts["user_name"]
        if facts.get("user_background"):
            feynman_instance.long_memory["user_background"] = facts["user_background"]
        if facts.get("new_fact"):
            feynman_instance.long_memory["interesting_facts_about_user"].append(facts["new_fact"])
        save_memory(feynman_instance.long_memory)
    except Exception as e:
        pass  # memory update failing shouldn't break the chat

def run():
    print("=" * 50)
    print("🧠 FEYNMAN DIGITAL TWIN")
    print("Type 'quit' to exit | 'memory' to see what Feynman remembers")
    print("=" * 50 + "\n")

    while True:
        user_input = input("You: ").strip()

        if not user_input:
            continue

        if user_input.lower() == "quit":
            print("Feynman: Well, it's been a pleasure! Keep asking questions!")
            break

        if user_input.lower() == "memory":
            print("\n📂 Current long-term memory:")
            print(json.dumps(feynman.long_memory, indent=2))
            print()
            continue

        reply = safe_chat(feynman, user_input)
        if reply:
            print(f"\nFeynman: {reply}\n")
            extract_and_save_memory(feynman, user_input)

run()
````
